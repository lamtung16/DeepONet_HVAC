```python
import pandas as pd
import numpy as np
import csv
```

```python
start = 512
end   = 1024

src_file = 'HVAC_B90_102_exp_10m_20210424.csv'
df = pd.read_csv(src_file)
Tz_k   = np.array(df['room_temp'])[start: end]
Tsa_k1 = np.array(df['supply_discharge_temp'])[start+1: end+1]
msa_k1 = np.array(df['airflow_current'])[start+1: end+1]
```

```python
data = np.zeros((end - start)*3)
for i in range(end - start):
    data[2*i] = Tsa_k1[i]
    data[2*i+1] = msa_k1[i]
    data[2*(end-start)+i] = Tz_k[i]
```

```python
# creat a new file data.csv
f = open('data.csv', 'w')
f.close()
```

```python
with open('data.csv', 'a', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(data)
```
